```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
from matplotlib.patches import Circle, Rectangle, Arc

def draw_rink(ax=None, color='black', lw=2, outer_lines=False):
    if ax is None:
        ax = plt.gca()
    
    outer_box_1 = Rectangle((-1500, 3000), 1485, 0, linewidth=lw, color='red')
    outer_box_2 = Rectangle((1500, 3000), -1485, 0, linewidth=lw, color='red')
    outer_box_3 = Rectangle((-700, 6000), 1500, 0, linewidth=lw, color=color)
    outer_box_4 = Rectangle((-1500, 3000), 0, 2200, linewidth=lw, color=color)
    outer_box_5 = Rectangle((1500, 3000), 0, 2200, linewidth=lw, color=color)
    blue_spot = Circle((0, 3000), radius=15, linewidth=lw, color='blue', fill=True)
    outer_arc_left = Arc((-700, 5200), 1600, 1600, theta1=90, theta2=180, linewidth=lw, color=color)
    outer_arc_right = Arc((700, 5200), 1600, 1600, theta1=0, theta2=90, linewidth=lw, color=color)
    
    center_outer_arc = Arc((0, 3000), 450, 450, theta1=0, theta2=180,
                           linewidth=lw, color='blue')
    
    red_spot_1 = Circle((-700, 3564), radius=30, linewidth=lw, color='red', fill=True)
    
    red_spot_2 = Circle((700, 3564), radius=30, linewidth=lw, color='red', fill=True)
    
    red_spot_3 = Circle((-700, 5000), radius=30, linewidth=lw, color='red', fill=True)
    red_spot_3_circle = Circle((-700, 5000), radius=450, linewidth=lw, color='red', fill=False)
    red_spot_3_outer_line_1 = Rectangle((-250, 4915), 60, 5, linewidth=lw, color='red')
    red_spot_3_outer_line_2 = Rectangle((-250, 5085), 60, 5, linewidth=lw, color='red')
    red_spot_3_outer_line_3 = Rectangle((-1150, 4915), -60, 5, linewidth=lw, color='red')
    red_spot_3_outer_line_4 = Rectangle((-1150, 5085), -60, 5, linewidth=lw, color='red')
    red_spot_3_inter_line_1 = Rectangle((-722.5, 5060), -60, 5, linewidth=lw, color='red')
    red_spot_3_inter_line_2 = Rectangle((-722.5, 5060), 5, 60, linewidth=lw, color='red')
    red_spot_3_inter_line_3 = Rectangle((-677.5, 5060), 60, 5, linewidth=lw, color='red')
    red_spot_3_inter_line_4 = Rectangle((-677.5, 5060), 5, 60, linewidth=lw, color='red')
    red_spot_3_inter_line_5 = Rectangle((-722.5, 4940), -60, 5, linewidth=lw, color='red')
    red_spot_3_inter_line_6 = Rectangle((-722.5, 4940), 5, -60, linewidth=lw, color='red')
    red_spot_3_inter_line_7 = Rectangle((-677.5, 4940), 60, 5, linewidth=lw, color='red')
    red_spot_3_inter_line_8 = Rectangle((-677.5, 4940), 5, -60, linewidth=lw, color='red')
    
    red_spot_4 = Circle((700, 5000), radius=30, linewidth=lw, color='red', fill=True)
    red_spot_4_circle = Circle((700, 5000), radius=450, linewidth=lw, color='red', fill=False)
    red_spot_4_outer_line_1 = Rectangle((250, 4915), -60, 5, linewidth=lw, color='red')
    red_spot_4_outer_line_2 = Rectangle((250, 5085), -60, 5, linewidth=lw, color='red')
    red_spot_4_outer_line_3 = Rectangle((1150, 4915), 60, 5, linewidth=lw, color='red')
    red_spot_4_outer_line_4 = Rectangle((1150, 5085), 60, 5, linewidth=lw, color='red')
    red_spot_4_inter_line_1 = Rectangle((677.5, 5060), -60, 5, linewidth=lw, color='red')
    red_spot_4_inter_line_2 = Rectangle((677.5, 5060), 5, 60, linewidth=lw, color='red')
    red_spot_4_inter_line_3 = Rectangle((722.5, 5060), 60, 5, linewidth=lw, color='red')
    red_spot_4_inter_line_4 = Rectangle((722.5, 5060), 5, 60, linewidth=lw, color='red')
    red_spot_4_inter_line_5 = Rectangle((677.5, 4940), -60, 5, linewidth=lw, color='red')
    red_spot_4_inter_line_6 = Rectangle((677.5, 4940), 5, -60, linewidth=lw, color='red')
    red_spot_4_inter_line_7 = Rectangle((722.5, 4940), 60, 5, linewidth=lw, color='red')
    red_spot_4_inter_line_8 = Rectangle((722.5, 4940), 5, -60, linewidth=lw, color='red')
    
    goal_line_1 = Rectangle((-122, 5600), 5, -137, linewidth=lw, color='red')
    
    goal_line_2 = Rectangle((122, 5600), 5, -137, linewidth=lw, color='red')
    
    goal_arc = Arc((0, 5463), 244, 46, theta1=183, theta2=0, linewidth=lw, color='red')

    blue_line = Rectangle((-1500, 3714), 3000, 30, linewidth=lw, color='blue')

    red_line = Rectangle((-1400, 5600), 2800, 5, linewidth=lw, color='red')

    
    rink_elements = [outer_box_1, outer_box_2, outer_box_3, outer_box_4, outer_box_5, blue_spot, outer_arc_left, outer_arc_right,
                     center_outer_arc, red_spot_1, red_spot_2, red_spot_3, red_spot_4,
                     goal_line_1, goal_line_2, goal_arc, blue_line, red_line, 
                     red_spot_3_circle, red_spot_3_outer_line_1, red_spot_3_outer_line_2, red_spot_3_outer_line_3, red_spot_3_outer_line_4, red_spot_3_inter_line_1, red_spot_3_inter_line_2, red_spot_3_inter_line_3, red_spot_3_inter_line_4, red_spot_3_inter_line_5, red_spot_3_inter_line_6, red_spot_3_inter_line_7, red_spot_3_inter_line_8,
                     red_spot_4_circle, red_spot_4_outer_line_1, red_spot_4_outer_line_2, red_spot_4_outer_line_3, red_spot_4_outer_line_4, red_spot_4_inter_line_1, red_spot_4_inter_line_2, red_spot_4_inter_line_3, red_spot_4_inter_line_4, red_spot_4_inter_line_5, red_spot_4_inter_line_6, red_spot_4_inter_line_7, red_spot_4_inter_line_8]

    for element in rink_elements:
        ax.add_patch(element)
        
    return ax
```

```python
plt.figure(figsize=(12,11))
draw_rink(outer_lines=True)
plt.xlim(-2000,2000)
plt.ylim(2500,6500)
plt.show()
```

```python
Data = pd.read_csv('C://Users/Pierre/Documents/Python Scripts/RHE76/shot_plot_game_2.csv')
Amiens = Data[Data.TEAM == 'AMIENS']
Grenoble = Data[Data.TEAM == 'GRENOBLE']
```

```python
plt.figure(figsize=(12,11))
plt.scatter(1500*Grenoble.LOC_Y/1300, 6000*(Grenoble.LOC_X+3000)/6050)
draw_rink(outer_lines=True)
plt.xlim(-2000,2000)
plt.ylim(2500,6500)
plt.show()
```

```python
joint_shot_chart = sns.jointplot((1500*Amiens.LOC_Y/1300),(6000*(Amiens.LOC_X+3000)/6050), stat_func=None,
                                 kind='scatter', space=0, alpha=0.5)
joint_shot_chart.fig.set_size_inches(12,11)
ax = joint_shot_chart.ax_joint
draw_rink(ax)
ax.set_xlim(-2000,2000)
ax.set_ylim(2500, 6500)
ax.set_xlabel('')
ax.set_ylabel('')
ax.tick_params(labelbottom='off', labelleft='off')
ax.set_title('Match 2 Demi-finale Amiens', y=1.2, fontsize=18)
plt.show()
```
